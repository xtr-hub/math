# 熵权法

**原理**：基于信息熵计算各指标的客观权重。指标值的差异越大 → 信息量越大 → 权重越高。

**步骤**：正向化 → 列和归一化得比重矩阵 → 计算信息熵 → 差异系数 → 归一化得权重


```python
import numpy as np
import sys; sys.path.insert(0, '..')
from src.algorithms.entropy_weight import calculate_entropy_weights
from src.utils.plot import Plotter
```

## 示例：学生综合评价

5 名学生 × 4 项指标（均为极大型），用熵权法自动确定各科权重。


```python
# 5 名学生 × 4 科成绩
matrix = np.array([
    [85.0, 90.0, 78.0, 92.0],
    [76.0, 88.0, 85.0, 85.0],
    [90.0, 82.0, 92.0, 78.0],
    [70.0, 95.0, 80.0, 88.0],
    [82.0, 78.0, 88.0, 90.0],
])

students = [f"学生{i+1}" for i in range(5)]
ind_labels = ["语文", "数学", "英语", "综合"]
```

```python
# 计算熵权（全部为极大型指标）
weights = calculate_entropy_weights(matrix)

print("熵权法确定的各指标权重：")
for name, w in zip(ind_labels, weights):
    bar = "█" * int(w * 50)
    print(f"  {name}：{w:.4f}  {bar}")
print(f"\n权重之和：{np.sum(weights):.4f}")
```

```python
# 混合指标类型：语文=极大型，数学=极小型，英语=极大型
matrix2 = np.array([
    [85.0, 70.0, 78.0],
    [76.0, 80.0, 85.0],
    [90.0, 75.0, 92.0],
    [70.0, 85.0, 80.0],
])

# kinds=[1, 2, 1] 指定第二列为极小型
weights2 = calculate_entropy_weights(matrix2, kinds=[1, 2, 1])
print("混合指标类型下的权重：")
for name, w in zip(["语文(大型)", "数学(小型)", "英语(大型)"], weights2):
    print(f"  {name}：{w:.4f}")
```

```python
# 可视化
fig, axes = Plotter.subplots(1, 2, figsize=(13, 5))

Plotter.heatmap(matrix, row_labels=students, col_labels=ind_labels,
                ax=axes[0], title="原始成绩矩阵", fmt=".0f")
Plotter.bar(ind_labels, weights, ax=axes[1], title="熵权法指标权重")

Plotter.save("../data/entropy_weight_result.png")
print("图片已保存至 data/entropy_weight_result.png")
```
